```python
import pandas as pd
import matplotlib.pyplot as plt 
import matplotlib
from matplotlib import font_manager, rc
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
d_bank = pd.read_csv("C:/Users/82104/Desktop/study/기업자금조달현황(은행대출).csv", encoding="cp949")
d_person = pd.read_csv("C:/Users/82104/Desktop/study/기업자금조달현황(직접금융).csv", encoding="cp949")
d_stock = pd.read_csv("C:/Users/82104/Desktop/study/등록 상장회사 증감률.csv", encoding="cp949")
```

```python
display(d_bank)
display(d_person)
display(d_stock)
```

```python
d_bank['날짜'] = pd.to_datetime(d_bank['날짜'], format='%Y-%m')
d_person['날짜'] = pd.to_datetime(d_person['날짜'], format='%Y-%m')
d_stock['날짜'] = pd.to_datetime(d_stock['날짜'], format='%Y-%m')
```

```python
d_bank = d_bank.pivot_table(index='날짜', values=['기업대출','대기업','중소기업'])
d_person = d_person.pivot_table(index='날짜', values=['직접금융','대기업','중소기업','금융채','ABS'])
d_stock = d_stock.pivot_table(index='날짜', values=['합계','유가증권시장','코스닥시장'])
```

```python
start_date = d_bank.index.min() - pd.DateOffset(day=1)
end_date = d_bank.index.max() + pd.DateOffset(day=31)
start_date = d_person.index.min() - pd.DateOffset(day=1)
end_date = d_person.index.max() + pd.DateOffset(day=31)
start_date = d_stock.index.min() - pd.DateOffset(day=1)
end_date = d_stock.index.max() + pd.DateOffset(day=31)
```

```python
dates = pd.date_range(start_date, end_date, freq='D')
dates.name = '날짜'
d_bank = d_bank.reindex(dates, method='ffill')
dates = pd.date_range(start_date, end_date, freq='D')
dates.name = '날짜'
d_person = d_person.reindex(dates, method='ffill')
dates = pd.date_range(start_date, end_date, freq='D')
dates.name = '날짜'
d_stock = d_stock.reindex(dates, method='ffill')
```
